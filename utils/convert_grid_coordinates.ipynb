```python
from pyproj import Transformer
import numpy as np

from grid_and_bathy import get_grid_angle, rotate_grid, translate_grid
from configs.config_object import ConfigObject
```

```python
config = ConfigObject('../config.json')
```

# Get MITgcm grid characteristics

```python
crs_target = '2056'
config_coord_converter = Transformer.from_crs(f"EPSG:{config.epsg_projection}", f"EPSG:{crs_target}", always_xy=True)
```

```python
(x0, y0) = config_coord_converter.transform(config.x0, config.y0)
print(x0, y0)
```

```python
rot_angle_in_degrees = config.rotation
rot_angle_in_radian = np.deg2rad(rot_angle_in_degrees)
print(rot_angle_in_degrees)
```

# Convert coordinates point into MITgcm grid coordinates

```python
y_point, x_point = 46.904, 6.843
point_epsg_projection = '4326'
```

```python
point_coord_converter = Transformer.from_crs(f"EPSG:{point_epsg_projection}", f"EPSG:{crs_target}", always_xy=True)
x,y = point_coord_converter.transform(x_point,y_point)
print(x,y)
```

```python
y - y0
```

```python
x_trans, y_trans = translate_grid(x, y, -x0, -y0)
print(x_trans, y_trans)
```

```python
x_mitgcm, y_mitgcm = rotate_grid(x_trans, y_trans, 0, 0, -rot_angle_in_radian)
print(f"{x_mitgcm}, {y_mitgcm}")
```
